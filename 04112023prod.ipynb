#import

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier

import pickle
```

#Загрузка данных

```python
!git clone https://github.com/Lasttrader/DA_repo
```

```python
path = '/content/DA_repo/titanic_train.csv'
```

```python
data = pd.read_csv(path, index_col = 0)
data.head()
```

```python
data['Embarked'].value_counts()
```

#Обработка категориальных признаков

```python
cat_cols = data.describe(include='object').columns.to_list()
```

```python
cat_cols[1:]
```

```python
data_cat = data[cat_cols[1:]]
data_cat.head()
```

##labelEncoder

```python
le = LabelEncoder()
for col in list(data_cat.columns):
  data_cat[col + '_LE'] = le.fit_transform(data_cat[col])
  with open(f'{col}_LE.pkl', 'wb') as f:
    pickle.dump(le, f) #сохранение модели кодировщика в файл
data_cat.head()
```

```python
data_cat_final = data_cat.drop(cat_cols[1:], axis =1)
data_cat_final.head()
```

```python
data_cat_final.isna().sum()
```

#Числовые признаки

```python
num_cols = data.describe().columns.to_list()
data_num = data[num_cols]
data_num.head()
```

#Нормализация

```python
'''
1. открытие приложения
2. вводит категориальные признаки
3. вводить числовые признаки
4. категориальные призхнаки  в индексы (кодирование)
5. применить standardscaler на всё векторе
6. predict
7. result
'''
```

```python
dataset = pd.concat([data_num, data_cat_final], axis =1)
dataset.head()
```

##StandardScaler

```python
scaler = StandardScaler()
dataset_scaled = scaler.fit_transform(dataset.drop('Survived', axis =1))
```

```python
with open('num_scaler.pkl', 'wb') as f:
  pickle.dump(scaler, f) #сохраняем скелер, для приложения
```

```python
dataset_scaled_df = pd.DataFrame(data = dataset_scaled,
                                 columns = dataset.drop('Survived', axis =1).columns.to_list())
dataset_scaled_df.head()
```

```python
dataset_scaled_df['Age'] = dataset_scaled_df['Age'].fillna(dataset_scaled_df['Age'].mean())
```

```python
dataset_scaled_df.isna().sum()
```

#X,y train_test_split

```python
y = np.array(data['Survived'])
X = np.array(dataset_scaled_df)

X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                    test_size = 0.2,
                                                    shuffle = True,
                                                    random_state = 42)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

#ML

```python
kNN = KNeighborsClassifier(n_neighbors=5)
kNN.fit(X_train, y_train)
```

```python
kNN.predict(X_test[:5])
```

```python
y_test[:5]
```

##Сохраняем модель ML

```python
with open('kNN.pkl', 'wb') as f:
  pickle.dump(kNN, f) #сохраняем скелер, для приложения
```
